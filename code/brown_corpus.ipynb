```python
from nltk.corpus import brown 
from collections import defaultdict
from math import log
from nltk.stem import *
from nltk.stem.porter import *

STEMMER = PorterStemmer()

total_counts = defaultdict(int)
tf = defaultdict(int)
df = defaultdict(int)

def preprocess_word(word):
    word = word.lower() #remove capital letters
    return STEMMER.stem(word)


doc_num = 0
for cat in brown.categories():
    cat_ids = brown.fileids(categories=cat)
    for id in cat_ids:
        sequence_of_words = brown.words(id)
        sequence_of_words = [preprocess_word(word) for word in sequence_of_words]
        for word in sequence_of_words:
            tf[(word,doc_num)] += 1
            total_counts[word] += 1
        
        word_set = set(sequence_of_words)
        for word in word_set:
            df[word] += 1

        print('Finished ', doc_num)
        doc_num += 1

w = defaultdict(list)

for term in df.keys():
    tf_idf_vector = []
    for doc in range(doc_num):
        tf_idf_vector.append(log(tf[(term, doc)]+1,10)*log((doc_num+1)/df[term]))
    w[term] = tf_idf_vector  #list of tf-idf vectors
```

```python
import numpy as np

def cosine_sim(vec1, vec2):
    return np.dot(vec1,vec2)/(np.sum(np.array(vec1)**2)*np.sum(np.array(vec2)**2))**(0.5)

test_words = ['politics','history','science','radical']
for word in test_words:
    if preprocess_word(word) not in w.keys():
        print(word, ' not in dictionary')

all_pairs = [(x,y) for x in test_words for y in test_words if x < y]

for word_pair in all_pairs:
    word1 = preprocess_word(word_pair[0])
    word2 = preprocess_word(word_pair[1])
    vec1 = w[word1]
    vec2 = w[word2]
    print(word_pair[0], word_pair[1], cosine_sim(vec1, vec2))
```
